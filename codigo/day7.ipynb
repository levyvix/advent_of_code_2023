```python
cartas_em_ordem_de_forca = [
    "A",
    "K",
    "Q",
    "J",
    "T",
    "9",
    "8",
    "7",
    "6",
    "5",
    "4",
    "3",
    "2",
][::-1]


def carta_para_numero(carta):
    if carta in cartas_em_ordem_de_forca:
        return cartas_em_ordem_de_forca.index(carta)
    else:
        raise ValueError("Carta inválida")


tipos_mapping = {
    "five of a kind": 7,
    "four of a kind": 6,
    "full house": 5,
    "three of a kind": 4,
    "two pairs": 3,
    "one pair": 2,
    "high card": 1,
}

# maos, bids = zip(*[line.split() for line in example.splitlines()])
maos, bids = zip(*[line.split() for line in data.splitlines()])
bids = list(map(int, bids))
maos_value = [[carta_para_numero(carta) for carta in mao] for mao in maos]
tipos = []

for m in maos:
    m_counter = Counter(m)

    if 5 in m_counter.values():
        tipos.append("five of a kind")
    elif 4 in m_counter.values():
        tipos.append("four of a kind")
    elif 3 in m_counter.values() and 2 in m_counter.values():
        tipos.append("full house")
    elif 3 in m_counter.values():
        tipos.append("three of a kind")
    elif 2 in m_counter.values():
        if len(m_counter) == 3:
            tipos.append("two pairs")
        else:
            tipos.append("one pair")
    else:
        tipos.append("high card")

lista_pontos = [tipos_mapping[tipo] for tipo in tipos]
indices_pontos = list(zip(range(len(maos)), lista_pontos))

lista_pontos_sorted = sorted(indices_pontos, key=lambda x: x[1], reverse=True)
lista_pontos_sorted = list(map(list, lista_pontos_sorted))

indices_pontos_sorted = sorted(indices_pontos, key=lambda x: (x[1], maos_value[x[0]]), reverse=True)

# Criar os rankings
rankings = [i for i, _ in indices_pontos_sorted]

rankings

# print(lista_pontos_sorted)

# rankings = []

# meta_i = 0
# while len(rankings) < len(lista_pontos_sorted):
#     i, p = lista_pontos_sorted[meta_i]
#     if i in rankings:
#         continue

#     if meta_i == len(lista_pontos_sorted) - 1:
#         rankings.append(i)
#         break

#     print(
#         f"now: {i, p}, after: {lista_pontos_sorted[meta_i+1][0], lista_pontos_sorted[meta_i+1][1]}"
#     )
#     if p == lista_pontos_sorted[meta_i + 1][1]:
#         print("empate")
#         #  resolve empate
#         for carta in range(5):
#             print(f"mao de agora: {maos[i]} -> {maos[i][carta]}")
#             print(
#                 f"mao de depois: {maos[lista_pontos_sorted[meta_i+1][0]]} -> {maos[lista_pontos_sorted[meta_i+1][0]][carta]}"
#             )
#             if (
#                 maos_value[i][carta]
#                 > maos_value[lista_pontos_sorted[meta_i + 1][0]][carta]
#             ):
#                 print("mao ganhadora", i)
#                 rankings.append(i)
#                 break

#             elif (
#                 maos_value[i][carta]
#                 < maos_value[lista_pontos_sorted[meta_i + 1][0]][carta]
#             ):
#                 rankings.append(lista_pontos_sorted[meta_i + 1][0])
#                 print("mao ganhadora", lista_pontos_sorted[meta_i + 1][0])
#                 lista_pontos_sorted[meta_i + 1], lista_pontos_sorted[meta_i] = (
#                     lista_pontos_sorted[meta_i],
#                     lista_pontos_sorted[meta_i + 1],
#                 )
#                 break
#         # else:  # All cards are exactly the same
#         #     print("Unresolved tie", i, lista_pontos_sorted[meta_i + 1][0])
#         #     rankings.append(i)
#         #     rankings.append(lista_pontos_sorted[meta_i + 1][0])
#     else:
#         rankings.append(i)

#     meta_i += 1


# rankings
```
