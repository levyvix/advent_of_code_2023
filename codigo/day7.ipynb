```python
cartas_em_ordem_de_forca = [
    "A",
    "K",
    "Q",
    "J",
    "T",
    "9",
    "8",
    "7",
    "6",
    "5",
    "4",
    "3",
    "2",
][::-1]


def carta_para_numero(carta):
    if carta in cartas_em_ordem_de_forca:
        return cartas_em_ordem_de_forca.index(carta)
    else:
        raise ValueError("Carta inválida")


tipos_mapping = {
    "five of a kind": 7,
    "four of a kind": 6,
    "full house": 5,
    "three of a kind": 4,
    "two pairs": 3,
    "one pair": 2,
    "high card": 1,
}

# maos, bids = zip(*[line.split() for line in example.splitlines()])
maos, bids = zip(*[line.split() for line in data.splitlines()])
bids = list(map(int, bids))
maos_value = [[carta_para_numero(carta) for carta in mao] for mao in maos]
tipos = []

for m in maos:
    m_counter = Counter(m)

    if 5 in m_counter.values():
        tipos.append("five of a kind")
    elif 4 in m_counter.values():
        tipos.append("four of a kind")
    elif 3 in m_counter.values() and 2 in m_counter.values():
        tipos.append("full house")
    elif 3 in m_counter.values():
        tipos.append("three of a kind")
    elif 2 in m_counter.values():
        if len(m_counter) == 3:
            tipos.append("two pairs")
        else:
            tipos.append("one pair")
    else:
        tipos.append("high card")

lista_pontos = [tipos_mapping[tipo] for tipo in tipos]
indices_pontos = list(zip(range(len(maos)), lista_pontos))

lista_pontos_sorted = sorted(indices_pontos, key=lambda x: x[1], reverse=True)
lista_pontos_sorted = list(map(list, lista_pontos_sorted))

indices_pontos_sorted = sorted(
    indices_pontos, key=lambda x: (x[1], maos_value[x[0]]), reverse=True
)

# Criar os rankings
rankings = [i for i, _ in indices_pontos_sorted]

rankings
```

```python
sums = 0
for i, r in enumerate(rankings[::-1]):
    print(
        f"RANK {i+1}: {maos[r]} -> type: {tipos[r]} -> bid: {bids[r]} -> rank*bid: {bids[r] * (i + 1)} -> total: {sums}"
    )
    sums += bids[r] * (i + 1)
sums
```

```python
cartas_em_ordem_de_forca = [
    "A",
    "K",
    "Q",
    "T",
    "9",
    "8",
    "7",
    "6",
    "5",
    "4",
    "3",
    "2",
    "J",
][::-1]


def carta_para_numero(carta):
    if carta in cartas_em_ordem_de_forca:
        return cartas_em_ordem_de_forca.index(carta)
    else:
        raise ValueError("Carta inválida")


tipos_mapping = {
    "five of a kind": 7,
    "four of a kind": 6,
    "full house": 5,
    "three of a kind": 4,
    "two pairs": 3,
    "one pair": 2,
    "high card": 1,
}

# maos, bids = zip(*[line.split() for line in example.splitlines()])
maos, bids = zip(*[line.split() for line in data.splitlines()])
bids = list(map(int, bids))
maos = list(maos)
maos_best = maos.copy()
# maos_value = [[carta_para_numero(carta) for carta in mao] for mao in maos]
tipos = []


def calculate_points(hand: str = "AAAAA") -> Literal[1, 2, 3, 4, 5, 6, 7]:
    hand_counter = Counter(hand)

    if 5 in hand_counter.values():
        return 7
    elif 4 in hand_counter.values():
        return 6
    elif 3 in hand_counter.values() and 2 in hand_counter.values():
        return 5
    elif 3 in hand_counter.values():
        return 4
    elif 2 in hand_counter.values():
        if len(hand_counter) == 3:
            return 3
        else:
            return 2
    else:
        return 1


possible_cards = ["2", "3", "4", "5", "6", "7", "8", "9", "T", "Q", "K", "A"][::-1]
for i, m in enumerate(maos):
    m_counter = Counter(m)

    if "J" in m_counter:
        max_points = -1
        best_card = None

        for card in possible_cards:
            temp_hand = [card if x == "J" else x for x in m]
            points = calculate_points(temp_hand)

            if points > max_points:
                max_points = points
                best_card = card

        maos_best[i] = "".join([best_card if x == "J" else x for x in m])
    else:
        maos_best[i] = m


for m in maos_best:
    m_counter = Counter(m)

    if 5 in m_counter.values():
        tipos.append("five of a kind")
    elif 4 in m_counter.values():
        tipos.append("four of a kind")
    elif 3 in m_counter.values() and 2 in m_counter.values():
        tipos.append("full house")
    elif 3 in m_counter.values():
        tipos.append("three of a kind")
    elif 2 in m_counter.values():
        if len(m_counter) == 3:
            tipos.append("two pairs")
        else:
            tipos.append("one pair")
    else:
        tipos.append("high card")


lista_pontos = [tipos_mapping[tipo] for tipo in tipos]  # string pra int - valor da mao
indices_pontos = list(zip(range(len(maos)), lista_pontos))  # (indice, pontos)
maos_value = [
    [carta_para_numero(carta) for carta in mao] for mao in maos
]  # string pra int - valor das cartas para comparacao uma a uma
indices_pontos_sorted = sorted(
    indices_pontos, key=lambda x: (x[1], maos_value[x[0]]), reverse=True
)  # ordena por pontos e depois por valor das cartas

rankings = [i for i, _ in indices_pontos_sorted]  # pega so os indices

sums = 0
for i, r in enumerate(rankings[::-1]):
    sums += bids[r] * (i + 1)
    print(
        f" INDEX: {r} \t -> RANK {i+1}: {maos[r]} \t -> type: {tipos[r]} \t -> bid: {bids[r]} -> rank*bid: {bids[r] * (i + 1)} -> total: {sums}"
    )
```
