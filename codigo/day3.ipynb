```python
def verifica_lados(start, end, line):
    symbols = ["@", "#", "$", "%", "&", "*", "/", "-", "=", "+"]

    if start == 0:
        return line[end] in symbols

    if end == len(line):
        return line[start - 1] in symbols

    if line[start - 1] in symbols:
        return True
    elif line[end] in symbols:
        return True
    return False


def verifica_baixo(start, end, line_abaixo):
    symbols = ["@", "#", "$", "%", "&", "*", "/", "-", "=", "+"]

    if start == 0:
        if any([char in symbols for char in line_abaixo[start : end + 1]]):
            return True
        else:
            return False

    if end == len(line_abaixo):
        if any([char in symbols for char in line_abaixo[start - 1 : end]]):
            return True
        else:
            return False

    return any([char in symbols for char in line_abaixo[start - 1 : end + 1]])


def verifica_acima(start, end, line_acima):
    symbols = ["@", "#", "$", "%", "&", "*", "/", "-", "=", "+"]

    if start == 0:
        if sum([char in symbols for char in line_acima[start : end + 1]]) > 0:
            return True
        else:
            return False

    if end == len(line_acima):
        if sum([char in symbols for char in line_acima[start - 1 : end]]) > 0:
            return True
        else:
            return False

    return sum([char in symbols for char in line_acima[start - 1 : end + 1]]) > 0


soma_numeros = 0
for i, linha in enumerate(data):
    # print("next line")
    for match in re.finditer(r"(\d+)", linha):
        start = match.start()
        end = match.end()

        # print(linha[start:end])

        lados = False
        acima = False
        abaixo = False

        if i == 0:
            abaixo = verifica_baixo(start, end, data[i + 1])
            lados = verifica_lados(start, end, linha)
        elif i == len(data) - 1:
            acima = verifica_acima(start, end, data[i - 1])
            lados = verifica_lados(start, end, linha)
        else:
            abaixo = verifica_baixo(start, end, data[i + 1])
            acima = verifica_acima(start, end, data[i - 1])
            lados = verifica_lados(start, end, linha)

        # print(
        #     f"abaixo: {abaixo}, acima: {acima}, lados: {lados}, overall: {abaixo or acima or lados}"
        # )

        if abaixo or acima or lados:
            soma_numeros += int(linha[start:end])


soma_numeros
```

```python
def verifica_acima_numeros(idx, linha_acima):
    numeros = []
    for number in re.finditer(r"(\d+)", linha_acima):
        start = number.start()
        end = number.end()

        n = list(range(start, end))

        ast = list(range(idx - 1, idx + 2, 1))

        intersect = list(set(n) & set(ast))

        if len(intersect) > 0:
            numeros.append(linha_acima[start:end])

    if numeros:
        return True, numeros
    return False, []


def verifica_abaixo_numeros(idx, linha_abaixo):
    numeros = []
    for number in re.finditer(r"(\d+)", linha_abaixo):
        start = number.start()
        end = number.end()

        n = list(range(start, end))

        ast = list(range(idx - 1, idx + 2, 1))

        intersect = list(set(n) & set(ast))

        if len(intersect) > 0:
            numeros.append(linha_abaixo[start:end])

    if numeros:
        return True, numeros
    return False, []


def verifica_lados_numeros(idx, linha_atual):
    numeros = []
    for number in re.finditer(r"(\d+)", linha_atual):
        start = number.start()
        end = number.end()

        if end - 1 == idx - 1 or start == idx + 1:
            numeros.append(linha_atual[start:end])

    if numeros:
        return True, numeros
    return False, []
```

```python
soma_numeros = 0
positive_numbers = []
for i, linha in enumerate(data):
    for match in re.finditer(r"(\*)", linha):
        start_ast = match.start()


        if i == 0:
            # primeira linha, so vê abaixo e lados
            tem_abaixo, numeros_abaixo = verifica_abaixo_numeros(start_ast, data[i + 1])
            tem_lados, numeros_lado = verifica_lados_numeros(start_ast, linha)

            if (tem_abaixo or tem_lados) and len(numeros_abaixo + numeros_lado) == 2:
                gear_ratio = math.prod([int(n) for n in numeros_abaixo + numeros_lado])
                positive_numbers.append(gear_ratio)
        if i == len(data) - 1:
            # ultima linha, so ve acima e lados

            tem_acima, numeros_acima = verifica_acima_numeros(start_ast, data[i - 1])
            tem_lados, numeros_lado = verifica_lados_numeros(start_ast, linha)
            if (tem_acima or tem_lados) and len(numeros_acima + numeros_lado) == 2:
                gear_ratio = math.prod([int(n) for n in numeros_acima + numeros_lado])
                positive_numbers.append(gear_ratio)
        else:
            tem_acima, numeros_acima = verifica_acima_numeros(start_ast, data[i - 1])
            tem_lados, numeros_lado = verifica_lados_numeros(start_ast, linha)
            tem_abaixo, numeros_abaixo = verifica_abaixo_numeros(start_ast, data[i + 1])

            if (
                (tem_abaixo
                or tem_acima
                or tem_lados)
                and len(numeros_abaixo + numeros_acima + numeros_lado) == 2
            ):
                gear_ratio = math.prod(
                    [int(n) for n in numeros_abaixo + numeros_acima + numeros_lado]
                )

                positive_numbers.append(gear_ratio)
```
